```python
import numpy as np
from skimage import feature, color, io
from sklearn import svm
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
import os
```

```python
def extract_hog_features(image):
    # gray_image = color.rgb2gray(image)
    hog_features = feature.hog(image, orientations=9, pixels_per_cell=(8, 8), cells_per_block=(2, 2), block_norm='L2-Hys')
    return hog_features
```

```python
positive_samples = []
path = "/home/wtsaichu/Dokumente/Arbeitsplatz/Dataset/Dataset/Dataset_V1.0/Binary/c/"
for file in os.listdir(path):
    image_path = path+file  # Change the path accordingly
    image = io.imread(image_path)
    hog_features = extract_hog_features(image)
    positive_samples.append(hog_features)
print(positive_samples)
```

```python
negative_samples = []
pathn = "/home/wtsaichu/Dokumente/Arbeitsplatz/Dataset/Dataset/Dataset_V1.0/Binary/I/"
for file in os.listdir(pathn):
    image_path = pathn+file  # Change the path accordingly
    image = io.imread(image_path)
    hog_features = extract_hog_features(image)
    negative_samples.append(hog_features)
pathn = "/home/wtsaichu/Dokumente/Arbeitsplatz/Dataset/Dataset/Dataset_V1.0/Binary/ok/"
for file in os.listdir(pathn):
    image_path = pathn+file  # Change the path accordingly
    image = io.imread(image_path)
    hog_features = extract_hog_features(image)
    negative_samples.append(hog_features)
```

```python
positive_labels = np.ones(len(positive_samples))
negative_labels = np.zeros(len(negative_samples))
```

```python
all_samples = np.concatenate((positive_samples, negative_samples), axis=0)
all_labels = np.concatenate((positive_labels, negative_labels), axis=0)
```

```python
X_train, X_test, y_train, y_test = train_test_split(all_samples, all_labels, test_size=0.2, random_state=42)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
clf = svm.SVC(kernel='sigmoid', C=2.0)
```

```python
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy * 100:.2f}%")
```
